```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# file path
path = '/Users/analeegraig/Documents/Semester 2/Machine Learning & Data Mining/Final Project/machine-learning-dse-i210-final-project-yelprestaurantrecommender/notebooks'
file_name = '/model_scores.csv'
```

```python
# read in data
test_scores = pd.read_csv(path + file_name)
test_scores = test_scores.drop('Unnamed: 0.1', axis = 1)
test_scores.loc[2, 'MAE Val'] = test_scores.loc[2, 'MAE Val']*-1
test_scores.loc[2, 'RMSE Val'] = test_scores.loc[2, 'RMSE Val']*-1
```

```python
# view data
test_scores
```

```python
# visualize RMSE
plt.figure(figsize = (10, 4))
bar_width = 0.4
bar_pos = np.arange(len(test_scores))

# plot bars
plt.barh(bar_pos - 0.5 * bar_width, test_scores['RMSE Test'], height = bar_width, align = 'center', label = 'RMSE Test')
plt.barh(bar_pos + 0.5 * bar_width, test_scores['RMSE Val'], height = bar_width, align = 'center', label = 'RMSE Val', color = '#202A44')

# add value labels
for i, value in enumerate(test_scores['RMSE Test']):
    plt.text(value, bar_pos[i] - 0.5 * bar_width, str(round(value, 2)), ha = 'left', va = 'center')
for i, value in enumerate(test_scores['RMSE Val']):
    plt.text(value, bar_pos[i] + 0.5 * bar_width, str(round(value, 2)), ha = 'left', va = 'center')

# make axes labels and title and show
plt.yticks(bar_pos, test_scores['Unnamed: 0'])
plt.xlabel('RMSE')
plt.title('Comparing Model Performance using RMSE')
plt.legend()
plt.show()
```

```python
# visualize MAE
plt.figure(figsize = (10, 4))
bar_width = 0.4
bar_pos = np.arange(len(test_scores))

# plot bars
plt.barh(bar_pos - 0.5 * bar_width, test_scores['MAE Test'], height = bar_width, align = 'center', label = 'MAE Test', color = '#6C935C')
plt.barh(bar_pos + 0.5 * bar_width, test_scores['MAE Val'], height = bar_width, align = 'center', label = 'MAE Val', color = '#2F4726')

# add value labels
for i, value in enumerate(test_scores['MAE Test']):
    plt.text(value, bar_pos[i] - 0.5 * bar_width, str(round(value, 2)), ha = 'left', va = 'center')
for i, value in enumerate(test_scores['MAE Val']):
    plt.text(value, bar_pos[i] + 0.5 * bar_width, str(round(value, 2)), ha = 'left', va = 'center')

# make axes labels and title and show
plt.yticks(bar_pos, test_scores['Unnamed: 0'])
plt.xlabel('MAE')
plt.title('Comparing Model Performance using MAE')
plt.legend()
plt.show()
```
