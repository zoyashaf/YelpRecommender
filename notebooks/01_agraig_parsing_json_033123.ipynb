````python
def parse_json(file_path):
    data = []
    with open(file_path) as f:
        for line in f:
            json_dict = json.loads(line)
            flattened_dict = pd.json_normalize(json_dict, sep = '_')
            data.append(flattened_dict)
    df = pd.concat(data, ignore_index = True)
    return df
````
